**Metrics**
|Relationship|Metric|
|------------|------|
|Equivalent  |	-   |
|Contains / Contained|Margins|
|Part	     |% of larger structure|
|Overlaps 	 |% of overlap|
|Surrounds   |Margins|

*Margins* 
- Max
- Min
- ***Orthogonal*** 


`bounds(geometry, **kwargs)`
Computes the bounds (extent) of a geometry.

For each geometry these 4 numbers are returned: 
> (min x, min y, max x, max y)
