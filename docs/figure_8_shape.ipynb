The equation for an ellipse centered at $(h, k)$ with total width $W$ (horizontal axis length) and total height $H$ (vertical axis length) is:

$\frac{(x - h)^2}{\left(\frac{W}{2}\right)^2} + \frac{(y - k)^2}{\left(\frac{H}{2}\right)^2} = 1$

If the ellipse is centered at the origin $(0,0)$, this simplifies to:

$\frac{x^2}{\left(\frac{W}{2}\right)^2} + \frac{y^2}{\left(\frac{H}{2}\right)^2} = 1$

Equivalently, using $a = W/2$ and $b = H/2$:

$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$

where $a$ is the semi-major axis length if $W \ge H$ (or semi-minor if $W < H$), and $b$ is the other semi-axis length.
I am looking for an equation for a shape made by combining parts of two circles a specified distance apart, with parts of  two ellipses a specified distance appart.  the ellipses should be located between the two circles with their long axis centered and perpendicular to a line between the centres of the two circles. the shape changes from being delimited by a circle type of equation to an ellipse type  of equation at  a point  where the circle and ellipse functions  are equal and the first derivatives of both functions are also equal

The shape you describe is a smooth composite curve consisting of two circular arcs (left and right ends) and two elliptical arcs (top and bottom) with continuous position and slope (C¹ continuity) at the junctions. Below I provide the implicit equations and the conditions that relate the geometry parameters.


## 1. Geometry setup (symmetric about both axes)

Let the two circles have equal radius $R$ and centers at $(-L, 0)$ and $(L, 0)$.  
The distance between circle centers is $2L$ (a specified value).  
The top and bottom elliptical arcs belong to ellipses whose major axes are perpendicular to the line joining the circle centers (i.e., vertical). Each ellipse is centered at $(0, \pm H_0)$ so that the top and bottom arcs are symmetric.

For the **top half** ($y \ge 0$), the boundary is made of:
- Left circular arc: from $x = -L - R$ to $x = -x_0$
- Central elliptical arc: from $x = -x_0$ to $x = x_0$
- Right circular arc: from $x = x_0$ to $x = L + R$

The join points $(\pm x_0, y_0)$ with $y_0 > 0$ are where the circle and ellipse meet tangently.


## 2. Equations of the pieces

**Circle** (right side, center $(L,0)$):

$(x - L)^2 + y^2 = R^2, \qquad y \ge 0$

**Ellipse** (top, centered at $(0, c)$ with $c > 0$):

$\frac{x^2}{a^2} + \frac{(y - c)^2}{b^2} = 1, \qquad y \ge c$

Here $a$ and $b$ are the semi-axes: $a$ (horizontal), $b$ (vertical).  
For a smooth top arc, the ellipse’s long axis is vertical ($b > a$) because it must be “taller” than it is wide to join the circles smoothly. However, the required orientation depends on the chosen parameters; the derivation below is general.


## 3. Smoothness conditions at the join point $(x_0, y_0)$

**Position continuity**:

$(x_0 - L)^2 + y_0^2 = R^2 \quad\text{(circle)}$

$\frac{x_0^2}{a^2} + \frac{(y_0 - c)^2}{b^2} = 1 \quad\text{(ellipse)}$

**Slope continuity** (differentiate implicitly):

- Circle: $\displaystyle y' = -\frac{x - L}{y}$
- Ellipse: $\displaystyle y' = -\frac{x\,b^2}{(y - c)\,a^2}$

Setting them equal at $(x_0, y_0)$:

$\frac{x_0 - L}{y_0} = \frac{x_0\,b^2}{(y_0 - c)\,a^2}$


## 4. Specifying the “distance between ellipses”

You mentioned “a specified distance apart” for the two ellipses. Most likely this means the **vertical distance between the top and bottom vertices** of the ellipses.  
For the top ellipse (center at $(0,c)$), the highest point is at $y = c + b$. For the bottom ellipse (center at $(0,-c)$), the lowest point is at $y = -(c + b)$. Thus the total height contributed by the elliptical parts is $2(c + b)$. The overall shape height is then $2(c + b)$ if the circles do not extend higher – but circles may have a different height if $R > c+b$. To avoid ambiguity, you can instead specify the **vertical distance between the two ellipse centers**, which is $2c$.


## 5. Solving for the join point and parameters

Given $L$ (half the circle-center distance) and $R$ (circle radius), and given a desired ellipse height parameter (say $b$ or $c$), equations (1) together with the two position equations determine $x_0, y_0, a$ (or another free parameter). In general, the system must be solved numerically. A convenient choice is to prescribe the ellipse’s **vertical semi-axis** $b$ and the center height $c$. Then (1) gives:

$a^2 = \frac{x_0\,b^2\,y_0}{(y_0 - c)(x_0 - L)}$

and the ellipse equation then yields $a$. The join point $x_0$ is found by requiring that the two $y_0$ expressions from circle and ellipse be equal – a root-finding problem.


## 6. Piecewise explicit equation for the top half

Once $x_0, y_0, a, b, c$ are known, the top boundary can be written as:

$y(x) =
\begin{cases}
\sqrt{R^2 - (x + L)^2}, & -L-R \le x \le -x_0 \\[4pt]
c + b\sqrt{1 - \frac{x^2}{a^2}}, & -x_0 \le x \le x_0 \\[4pt]
\sqrt{R^2 - (x - L)^2}, & x_0 \le x \le L+R
\end{cases}$

The bottom half is $y = -y(x)$ (by symmetry).


## 7. Example with specific numbers

Suppose you choose circle radius $R = 5$, half center distance $L = 4$ (so circles are 8 apart), and you want the ellipse center height $c = 2$ and vertical semi-axis $b = 3$. Then solving the continuity and tangency conditions yields approximately $x_0 \approx 2.78$ and $a \approx 4.12$. The resulting shape is smooth.


## 8. When you need a single equation

If you prefer a single implicit equation (nonsmooth at the joins, but continuous), you can use:

$\min\!\left( \frac{(x+L)^2 + y^2}{R^2},\; \frac{x^2}{a^2} + \frac{(y-c)^2}{b^2},\; \frac{(x-L)^2 + y^2}{R^2} \right) = 1$

with appropriate sign handling for the top/bottom halves. However, that is not an exact representation of the composite curve because the minimum function does not enforce the correct branch selection.


## Summary

The required shape is defined piecewise by circle and ellipse equations joined at points where both the coordinates and the first derivatives match. You must specify:
- Circle radius $R$ and center distance $2L$
- Ellipse parameters (e.g., vertical semi-axis $b$ and center offset $c$)
- Then solve for $x_0, y_0, a$ using the continuity and tangency equations.


I’ll provide Python code that numerically solves for the smooth join parameters and plots the composite shape.  


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# -------- Parameters --------
R = 5.0          # circle radius
L = 4.0          # half distance between circle centers (centers at ±L)
b_ell = 3.0      # vertical semi-axis of ellipse (b)
c_ell = 4.0      # y-coordinate of ellipse center

# -------- Solve for join point (x0, y0) and ellipse horizontal semi-axis a --------
def equations(vars):
    x0, y0, a = vars
    eq1 = (x0 - L)**2 + y0**2 - R**2
    eq2 = (x0**2)/(a**2) + ((y0 - c_ell)**2)/(b_ell**2) - 1.0
    # derivative matching: (x0-L)/y0 = (x0 * b_ell**2) / ((y0 - c_ell) * a**2)
    eq3 = (x0 - L) * (y0 - c_ell) * a**2 - (x0 * b_ell**2 * y0)
    return [eq1, eq2, eq3]

# Initial guess
x0_guess = 2.0
y0_guess = np.sqrt(R**2 - (x0_guess - L)**2)   # on circle
a_guess = 4.0
sol = fsolve(equations, [x0_guess, y0_guess, a_guess])
x0, y0, a = sol
print(f"Join point: x0 = {x0:.4f}, y0 = {y0:.4f}")
print(f"Ellipse a (horizontal semi-axis) = {a:.4f}")

# -------- Build the curve --------
# x ranges:
# Left circle: from -L-R to -x0
# Top ellipse: from -x0 to x0
# Right circle: from x0 to L+R

x_left = np.linspace(-L-R, -x0, 200)
y_left = np.sqrt(R**2 - (x_left + L)**2)

x_ell = np.linspace(-x0, x0, 300)
y_ell = c_ell + b_ell * np.sqrt(1 - (x_ell**2)/(a**2))

x_right = np.linspace(x0, L+R, 200)
y_right = np.sqrt(R**2 - (x_right - L)**2)

# Bottom half (symmetric)
x_left_b = np.linspace(-L-R, -x0, 200)
y_left_b = -np.sqrt(R**2 - (x_left_b + L)**2)

x_ell_b = np.linspace(-x0, x0, 300)
y_ell_b = -(c_ell + b_ell * np.sqrt(1 - (x_ell_b**2)/(a**2)))

x_right_b = np.linspace(x0, L+R, 200)
y_right_b = -np.sqrt(R**2 - (x_right_b - L)**2)

# -------- Plot --------
plt.figure(figsize=(8, 6))
plt.plot(x_left, y_left, 'b-', linewidth=2, label='Left circle')
plt.plot(x_ell, y_ell, 'r-', linewidth=2, label='Top ellipse')
plt.plot(x_right, y_right, 'b-', linewidth=2, label='Right circle')
plt.plot(x_left_b, y_left_b, 'b-', linewidth=2)
plt.plot(x_ell_b, y_ell_b, 'r-', linewidth=2)
plt.plot(x_right_b, y_right_b, 'b-', linewidth=2)

# Mark the join points
plt.plot([-x0, x0], [y0, y0], 'ko', markersize=5, label='Smooth joins')
plt.plot([-x0, x0], [-y0, -y0], 'ko', markersize=5)

# Add centers for reference
plt.plot([-L, L], [0, 0], 'k+', markersize=8, label='Circle centers')
plt.plot([0, 0], [c_ell, -c_ell], 'r+', markersize=8, label='Ellipse centers')

plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.6)
plt.title(f"Smooth composite shape: circles (R={R}, centers ±{L}) + ellipses (b={b_ell}, c={c_ell})")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
